# Univariate Bivariate and Multivariate Analysis

```python
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
import numpy as np

#import iris
iris = datasets.load_iris()
iris_df=pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df['Species']=iris.target

iris_df.head()
```

```python
#create a dataframe for target to rename the category

target_df=pd.DataFrame(iris.target)
#replace 0,1,2 label with species name
target_df.replace({0: 'Setosa',1:'Versicolour',2:'Virginica'},inplace=True)
#print to see change
target_df.head()
```

```python
#print the count of labels in each species-- frequency table'
iris_df['Species']=target_df
iris_df.tail(5)
```

```python
#change the column name without (cm) tag

iris_df.columns=['sepal length','sepal width', 'petal length','petal width','Species']
iris_df.head(5)
```

# Univariate Analysis 

“Uni” means one and “Variate” means variable hence univariate analysis means analysis of one variable or one feature. Univariate basically tells us how data in each feature is distributed and also tells us about central tendencies like mean, median, and mode.

```python
#descriptive statistics for numerical data
iris_df.describe()
```

```python
#categorical data
iris_df['Species'].value_counts()
```

```python
#categorical variable plot in a column
iris_df['Species'].value_counts().plot(kind='bar',color='red')
```

```python
#We use histogram to analyze frequency distribution and visualization
iris_df['sepal length'].hist()
```

```python
#box plot for value distribution
plt.boxplot(iris_df['sepal length'])
plt.show()
```

```python
#plot univariate values on x axis and y axis 0
df_setosa=iris_df.loc[iris_df['Species']=='Setosa']
df_virginica=iris_df.loc[iris_df['Species']=='Virginica']
df_versicolor=iris_df.loc[iris_df['Species']=='Versicolour']
df_setosa.head(5)

```

```python
#x axis is values for sepal length and y has no value so filling array with zeroes for y axis in shape of x data
plt.plot(df_setosa['sepal length'],np.zeros_like(df_setosa['sepal length']))
plt.plot(df_virginica['sepal length'],np.zeros_like(df_virginica['sepal length']))
plt.plot(df_versicolor['sepal length'],np.zeros_like(df_versicolor['sepal length']))
plt.xlabel('sepal length')
plt.show()
```

```python
#make plot little bigger by adding 'o'
plt.plot(df_setosa['sepal length'],np.zeros_like(df_setosa['sepal length']),'o')
plt.plot(df_virginica['sepal length'],np.zeros_like(df_virginica['sepal length']),'o')
plt.plot(df_versicolor['sepal length'],np.zeros_like(df_versicolor['sepal length']),'o')
plt.xlabel('sepal length')
plt.show()
```

# Bivariate Analysis
Two variables are compared to see their dependencies

```python
import seaborn as sns
```

```python
#plot with petal length and sepal width
sns.FacetGrid(iris_df, hue='Species',size=5).map(plt.scatter,'petal length', 'sepal width').add_legend()
```

## Multivariate analysis
Pairplot  be used to see pairwise feature analysis.  Heatmap can also be used to see interdependencies between features

```python
sns.pairplot(iris_df,hue='Species',height=3)
#the center column in output plot will be pdf 
```

#### Analysis
from graph it is observed petal length and petal width can be used for regression problem
